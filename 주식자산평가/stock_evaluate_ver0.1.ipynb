```python
import FinanceDataReader as fdr
import pandas as pd
```

### 미국 주식 불러오기

```python
code='AAPL'
stocks_df = pd.DataFrame()
tmp = fdr.DataReader(code, '2020')
tmp = tmp.rename(columns={'Close':code})
tmp
```

```python
code='MSFT'
tmp = fdr.DataReader(code, '2020')
tmp = tmp.rename(columns={'Close':code})
stocks_df = pd.concat([stocks_df, tmp[code]], axis=1)
stocks_df
```

### 여러 종목 불러와 df 만들기

```python
# 

stocks_list = ['AAPL', 'MSFT', 'TSLA', 'NVDA']
stocks_df = pd.DataFrame()
for code in stocks_list:
    tmp = fdr.DataReader(code, '2020')
    tmp = tmp.rename(columns={'Close':code})
    stocks_df = pd.concat([stocks_df, tmp[code]], axis=1)
stocks_df
```

### 국내 주식 불러오기

```python
# 셀트리온 : '068270'
# 삼성전자 : '005930'
# 인선이엔티 : '060150'
# 스튜디오드래곤 : '253450'
# KT : '030200'

stock_dic = {'068270':'셀트리온', '005930':'삼성전자', '060150':'인선이엔티', '253450':'스튜디오드래곤', '030200':'KT'}
stocks_df = pd.DataFrame()
for code in stock_dic.keys():
    name = stock_dic[code]
    tmp = fdr.DataReader(code, '2020')
    tmp = tmp.rename(columns={'Close':name})
    stocks_df = pd.concat([stocks_df, tmp[name]], axis=1)
stocks_df
```
